```python
#2.Handle Missing Values
```

```python
# Impute 'Age' with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Impute 'Embarked' with the mode
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace=True)

# Drop 'Cabin' due to excessive missing values
df.drop('Cabin', axis=1, inplace=True)

# Verify nulls are handled (excluding 'Cabin' which is dropped)
print("\n--- Nulls after Imputation ---")
print(df.isnull().sum())
```

```python
#3.Convert Categorical Features into Numerical (Encoding)
```

```python
# 1. 'Sex': Use Label Encoding (or simple mapping)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# 2. 'Embarked': Use One-Hot Encoding
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True) # drop_first avoids multicollinearity

print("\n--- Data after Encoding ---")
print(df[['Sex', 'Embarked_Q', 'Embarked_S']].head())
```

```python
#4.Normalize/Standardize Numerical Features
```

```python
from sklearn.preprocessing import StandardScaler

# Select features for scaling
features_to_scale = ['Age', 'Fare']
scaler = StandardScaler()

# Fit and transform
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])

print("\n--- Scaled Features (Age and Fare) ---")
print(df[['Age', 'Fare']].head())
```

```python
#5.Visualize Outliers and Remove Them
```

```python
# Visualize 'Fare' outliers
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['Fare'])
plt.title('Boxplot of Scaled Fare')
plt.show()


# Outlier Removal using IQR for 'Fare'
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_clean = df[(df['Fare'] >= lower_bound) & (df['Fare'] <= upper_bound)]

print(f"\nOriginal number of rows: {len(df)}")
print(f"Number of rows after outlier removal (Fare): {len(df_clean)}")
```

```python
#Exploratory Data Analysis (EDA)
```

```python
print("\n--- Summary Statistics of Clean Data ---")
print(df_clean.describe())
```

```python
#Create Histograms and Boxplots for Numeric Features
```

```python
# Set style
sns.set_style("whitegrid")

# Histogram for Pclass (as a pseudo-numeric discrete feature)
plt.figure(figsize=(6, 4))
sns.histplot(df_clean['Pclass'], bins=3, kde=False, discrete=True)
plt.title('Distribution of Passenger Class (Pclass)')
plt.show()


# Histogram for Scaled Age
plt.figure(figsize=(6, 4))
sns.histplot(df_clean['Age'], kde=True)
plt.title('Distribution of Scaled Age')
plt.show()
```

```python
# Use Pairplot/Correlation Matrix for Feature Relationships
# A correlation matrix is best for a quick overview of linear relationships.
```

```python
# Calculate the correlation matrix
correlation_matrix = df_clean.corr(numeric_only=True)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
# 4. & 5. Identify Patterns and Make Feature-Level Inferences (Survival Focus)
# Let's visualize the relationship between key features and the target variable, Survived.
```

```python
# Survival Rate by Pclass
plt.figure(figsize=(6, 4))
sns.barplot(x='Pclass', y='Survived', data=df_clean)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
